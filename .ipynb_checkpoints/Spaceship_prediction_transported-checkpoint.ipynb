Importing the Dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
```

Input the Data

```python
df = pd.read_csv('train-spaceship.csv')
df
```

Preprocessing the Data

```python
df_name_remove = df.drop(['Name'],axis=1)
df_name_remove
```

```python
df_name_remove.info()
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
df_name_remove['Transported'] = le.fit_transform(df_name_remove['Transported'])
df_name_remove.head()
```

```python
CAT_NUM_COL = [col for col in df_name_remove.columns if df_name_remove[col].dtype=='float64']
CAT_OBJ_COL = [col for col in df_name_remove.columns if df_name_remove[col].dtype=='object']
CAT_OBJ_COL
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
for data in df_name_remove[CAT_OBJ_COL].columns:
    df_name_remove[data] = le.fit_transform(df_name_remove[data])
```

```python
df_name_remove.head()
```

```python
df_name_remove['Tr']
```

```python
#nghien cuu du lieu
corr = df_name_remove.corr()
corr.Transported.sort_values(ascending=False)
```

```python
plt.figure(figsize=(9,9))
sns.heatmap(corr, annot=True)
```

```python
#split the data
from sklearn.model_selection import train_test_split
x = df_name_remove.drop(['Transported'],axis=1)
y = df_name_remove.Transported
print(x)
print(y)
```

```python
df_age_range = df_name_remove[(df_name_remove['Age'] >= 20) & (df_name_remove['Age'] <= 40)]

# Create a histogram using seaborn
sns.histplot(data=df_age_range, x='Age', bins=10, kde=False)
```

```python
#split data to prepare for training
Xtrain, Xtest, Ytrain, Ytest = train_test_split(x, y, test_size=0.75, random_state=10)
Xtrain.shape, Xtest.shape, Ytrain.shape, Ytest.shape
```

```python
Xtrain_no_missing = Xtrain.dropna()
Ytrain_no_missing = Ytrain[Xtrain.index.isin(Xtrain_no_missing.index)]

Xtrain_no_missing, Xtest, Ytrain_no_missing, Ytest = train_test_split(Xtrain_no_missing, Ytrain_no_missing, test_size=0.1, random_state=42)
#train model

from sklearn.ensemble import RandomForestClassifier
rtf = RandomForestClassifier()
rtf.fit(Xtrain_no_missing, Ytrain_no_missing)
```

```python
y_pred = rtf.predict(Xtest)
from sklearn.metrics import confusion_matrix, accuracy_score
accuracy_score(y_pred, Ytest)
```

```python
sns.heatmap(confusion_matrix(y_pred, Ytest),annot=True)
```

```python
print(len(y_pred))
```

```python
#standard scaler
X = df_name_remove.drop(['Transported'],axis=1)
Y = df_name_remove.Transported
X = (X-X.mean()) / X.std()
Xtrain, Xtest, Ytrain, Ytest = train_test_split(X, Y, test_size=0.105, random_state=42)
Xtrain.shape, Ytrain.shape, Xtest.shape, Ytest.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, accuracy_score

# Assuming Xtrain_no_missing and Ytrain_no_missing are defined
Xtrain_no_missing = Xtrain.dropna()
Ytrain_no_missing = Ytrain[Xtrain.index.isin(Xtrain_no_missing.index)]

# Adjust test_size to achieve the desired number of data points in the test set
Xtrain_no_missing, Xtest, Ytrain_no_missing, Ytest = train_test_split(
    Xtrain_no_missing, Ytrain_no_missing, test_size=4277/len(Ytrain_no_missing), random_state=42
)

# Train your model on Xtrain_no_missing, Ytrain_no_missing
rf = RandomForestClassifier()
rf.fit(Xtrain_no_missing, Ytrain_no_missing)

# Make predictions on the test set
y_pred_std = rf.predict(Xtest)

# Check the length of y_pred_std
print("Number of predicted values:", len(y_pred_std))
```

```python
accuracy_score(y_pred_std, Ytest)
```
